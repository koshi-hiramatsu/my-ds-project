```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 乱数でサンプル DataFrame を作成
np.random.seed(42)
df = pd.DataFrame({
    "Category": np.random.choice(["A","B","C"], size=50),
    "Value": np.random.randn(50).cumsum()
})
df.head()
```

```python
# カテゴリごとの平均と標準偏差
stats = df.groupby("Category")["Value"].agg(["mean","std"])
stats
```

```python
# 各カテゴリの累積値推移を折れ線グラフで可視化
plt.figure(figsize=(8,4))
for cat, sub in df.groupby("Category"):
    plt.plot(sub.index, sub["Value"], label=cat)
plt.legend(title="Category")
plt.title("Cumulative Value by Category")
plt.xlabel("Sample Index")
plt.ylabel("Cumulative Value")
plt.show()
```
